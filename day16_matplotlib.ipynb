```python

akbarData = []
arslanData = []
sub = ["Maths", "Computer", "Science", "Python","Sindhi", "Urdu", "Biology", "Java"]
for i in range(8):
    akbarMarks = int(input(f"Akbar: Enter Marks of Subject : {sub[i]}"))
    akbarData.append(akbarMarks)
    
for i in range(8):
    arslanMarks = int(input(f"Arslan: Enter Marks of Subject : {sub[i]}"))
    arslanData.append(arslanMarks)

# akbarData = [90, 80, 75, 91, 70, 85, 60, 73]
# arslanData = [17, 34,50, 5, 82, 55, 39, 41]


with open("marks.csv", "w") as f:
    f.write("Subject, Akbar, Arslan\n")
    for i in range(len(sub)):
        f.write(f"{sub[i]}, {akbarData[i]}, {arslanData[i]}\n")

```

```python
print(f"akbarData : {akbarData}")
print(f"arslanData : {arslanData}")
```

```python
data = pd.read_csv("marks.csv")
print(data)

plt.plot(data['Subject'], data[' Akbar'], 'b', label='Akbar')
plt.plot(data['Subject'], data[' Arslan'], 'r', label='Arslan')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Comparison between Akbar and Arslan')
plt.legend()

# plt.plot(sub, akbarData ,'b', sub, arslanData,'r')   
```

```python
totalMarks = 800
akbarObtainedMarks = np.sum(akbarData)
akbarPercentage = (akbarObtainedMarks/totalMarks)*100

arslanObtainedMarks = np.sum(arslanData)
arslanPercentage = (arslanObtainedMarks/totalMarks)*100

if akbarPercentage > 50:
    print(f"Akbar Passed, Percentage : {akbarPercentage}")
else:
    print(f"Akbar Failed, Percentage : {akbarPercentage}")

if arslanPercentage > 50:
    print(f"Arslan Passed, Percentage : {arslanPercentage}")
else:
    print(f"Arslan Failed , Percentage : {arslanPercentage}")
```
