```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error , r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
n = len(x)

# Calculate sums
sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum([x[i]*y[i] for i in range(n)])
sum_x2 = sum([x[i]**2 for i in range(n)])

# Calculate slope (m)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - (sum_x)**2)

# Calculate intercept (b)
b = (sum_y - m * sum_x) / n

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
y = m*(5)+b
y
```

```python
experience = np.array([1, 2,3,4,5]).reshape(-1,1)
salary = np.array([50000, 100000, 150000, 200000, 250000])
```

```python
model = LinearRegression()
```

```python
train = model.fit(experience,salary)
```

```python
y_pred = model.predict(experience)
y_pred
```

```python
MAE = mean_absolute_error(salary, y_pred)

print("MAE = ", MAE)
```

```python
MSE = mean_squared_error(salary, y_pred)
print("MSE = ", MSE)
```

```python
r2 = r2_score(salary, y_pred)
print("R2 = ", r2)
```

```python
plt.scatter(experience, salary, color="blue", label="Actual Data")
plt.plot(experience, y_pred, color="red", label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience Vs Salary")
plt.legend()
plt.show()
```
