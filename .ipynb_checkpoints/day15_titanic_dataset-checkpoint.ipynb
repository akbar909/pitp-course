```python
survival_counts = df['Survived'].value_counts()

print("Survival counts:")
survival_counts
```

```python
survived = survival_counts.get(1)
not_survived = survival_counts.get(0)

print(f"\nNumber of passengers who survived: {survived}")
print(f"Number of passengers who did not survive: {not_survived}")
```

```python
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()

survival_rate_by_gender*100
```

```python
average_age = df.groupby('Survived')['Age'].mean()

average_age
```

```python
port_counts = df['Embarked'].value_counts()
most_common_port = port_counts.idxmax()

print("All Ports" ,port_counts)
print("Most Common embarkation Post :" ,most_common_port)
```
